```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true', compression = 'zip')
```

```python
### Desafio 1: Por que a classe tratamento é tão desbalanceada? R: Pois o tratamento "com_controle" só é realizado com um tipo de droga:
ctrl = df.loc[df['tratamento'] == 'com_controle']
ctrl['droga'].unique()
```

```python
###Desafio 3: Qual é a proporção das classes tratamento? 
###Desafio 7: Deixar o gráfico bonitão:

colors = sns.color_palette('pastel')
plt.figure(figsize=(4, 3), tight_layout=True)
plt.pie(df['tratamento'].value_counts(), labels=df['tratamento'].unique(), autopct='%.0f %%', pctdistance=.7, colors=colors, shadow=True)
plt.title('Tratamentos', weight='bold')
plt.show()
```
